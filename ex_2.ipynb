```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import *

%matplotlib inline
```

# 1 вариант

```python
data = pd.read_csv('data_test.csv', encoding='ANSI')
```

```python
data
```

Сортируем в датафрейме столбцы каждой строки по алфавиту

```python
for i, row in enumerate(data.iterrows()):
    data.iloc[i, 1:5] = sorted(list(row[1][1:5]))

data = data.sort_values(['unit_1', 'unit_2', 'unit_3', 'unit_4'], axis=0, inplace=False)
```

Разбиваем на до и после

```python
data_before = data[data['balance_mark'] == 'before']
```

```python
data_after = data[data['balance_mark'] == 'after']
```

Аггрегируем по 4-м столбцам , получая среднее значение длинны боя

```python
mean_before = data_before.groupby(['unit_1', 'unit_2', 'unit_3', 'unit_4']).agg( \
    mean_len_before = pd.NamedAgg(column = 'len', aggfunc = 'mean'))
mean_before
```

```python
mean_after = data_after.groupby(['unit_1', 'unit_2', 'unit_3', 'unit_4']).agg( \
    mean_len_after = pd.NamedAgg(column = 'len', aggfunc = 'mean'))
mean_after
```

Объединяем

```python
pd_before_after = pd.concat([mean_before, mean_after], axis=1)
pd_before_after
```

Считаем разность, что стало после и что было до, т.к. ожидаем что настройки повлияли на длительность боя 

```python
pd_before_after['dif_len'] = pd_before_after['mean_len_after'] - pd_before_after['mean_len_before']
pd_before_after
```

```python
pd_before_after['dif_len'].describe()
```

# 2 Вариант

```python
data = pd.read_csv('data_test.csv', encoding='ANSI')
```

Разделим данные на before и after по balance_mark

```python
data_before = data[data['balance_mark'] == 'before']
```

```python
data_before
```

```python
data_after = data[data['balance_mark'] == 'after']
```

```python
data_after
```

Посмотрим на вариации значений

```python
data['unit_1'].value_counts()
```

посмотрим количество всех возможных комбинаций, где перестановка данных элементов нам не важна

```python
elements = ['меч', 'топор', 'булава', 'лук', 'посох']
unique_combinations = []
count = 0
for combination in combinations_with_replacement(elements, 4):
    unique_combinations.append(combination)
    count += 1
count
```

```python
results = []
for combination in unique_combinations:
    for sample in list(permutations(combination)):
        before = data_before[(data_before['unit_1'] == sample[0]) & (data_before['unit_2'] == sample[1]) \
            & (data_before['unit_3'] == sample[2]) & (data_before['unit_4'] == sample[3])]
        mean_len_before = before['len'].mean()
        
        after = data_after[(data_after['unit_1'] == sample[0]) & (data_after['unit_2'] == sample[1])\
           & (data_after['unit_3'] == sample[2]) & (data_after['unit_4'] == sample[3])]
        mean_len_after = after['len'].mean()
        result = [sample, mean_len_before, mean_len_after]
        if result in results:
            result = []
            continue
        else:
            results.append([sample, mean_len_before, mean_len_after])
        
```

Переведем все в таблицу

```python
pd_results = pd.DataFrame(columns=['unit_1', 'unit_2', 'unit_3', 'unit_4', 'len_before', 'len_after'])
for i in results:
    pd_results.loc[len(pd_results)] = [i[0][0], i[0][1], i[0][2], i[0][3], i[1], i[2]]
```

```python
pd_results
```

Добавим столбец, который показывает отличия между тем что было после и между тем что было до

```python
pd_results['dif_len'] = pd_results['len_after'] - pd_results['len_before']
```

```python
pd_results
```

```python
pd_results.describe()
```
